```python
import pandas as pd
import numpy as np
```

# 02 Load Dataset

```python
df_perolehan = pd.read_excel(r'C:\Users\hp\Internship\KPU Surabaya\Dataset\DPD\perolehan_suara_calon.xlsx')
df_pemilih = pd.read_excel(r'C:\Users\hp\Internship\KPU Surabaya\Dataset\DPD\pemilih_penggunaan_hak_pilih.xlsx')
df_surat = pd.read_excel(r'C:\Users\hp\Internship\KPU Surabaya\Dataset\DPD\penggunaan_surat_suara.xlsx')
df_disabilitas = pd.read_excel(r'C:\Users\hp\Internship\KPU Surabaya\Dataset\DPD\pemilih_disabilitas.xlsx')
```

# 03 Preprocessing

```python
# Melt data pemilih (id = 'uraian')
df_pemilih_melted = df_pemilih.melt(id_vars=['Uraian'], 
                                    var_name='Kecamatan', 
                                    value_name='Jumlah')
df_pemilih_melted.rename(columns={'uraian': 'Uraian'}, inplace=True)

# Melt data surat suara (id = 'Uraian')
df_surat_melted = df_surat.melt(id_vars=['Uraian'], 
                                var_name='Kecamatan', 
                                value_name='Jumlah')

# Melt data disabilitas (id = 'Uraian')
df_disabilitas_melted = df_disabilitas.melt(id_vars=['Uraian'], 
                                            var_name='Kecamatan', 
                                            value_name='Jumlah')

df_calon_only = df_perolehan.drop(columns=['Jumlah Suara Sah', 'Jumlah Suara Tidak Sah'])
```

```python
# Tentukan kolom nama calon (yaitu semua kolom kecuali kolom tetap)
kolom_tetap = ['No', 'Kecamatan', 'Jumlah Suara Sah', 'Jumlah Suara Tidak Sah']
kolom_calon = [col for col in df_perolehan.columns if col not in kolom_tetap]

# Melt hanya kolom nama calon (jadi 'Jumlah Suara Sah' dan 'Tidak Sah' tetap utuh)
df_calon_melted = df_perolehan.melt(
    id_vars=['No', 'Kecamatan', 'Jumlah Suara Sah', 'Jumlah Suara Tidak Sah'],
    value_vars=kolom_calon,
    var_name='nama_calon',
    value_name='jumlah_suara'
)

df_calon_melted.head()
```
