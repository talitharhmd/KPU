# 01 Import Library

```python
import pandas as pd
import numpy as np
```

# 02 Load Dataset

```python
df_perolehan = pd.read_excel(r'C:\Users\hp\Internship\KPU Surabaya\Dataset\DPD\perolehan_suara_calon.xlsx')
df_pemilih = pd.read_excel(r'C:\Users\hp\Internship\KPU Surabaya\Dataset\DPD\pemilih_penggunaan_hak_pilih.xlsx')
df_surat = pd.read_excel(r'C:\Users\hp\Internship\KPU Surabaya\Dataset\DPD\penggunaan_surat_suara.xlsx')
df_disabilitas = pd.read_excel(r'C:\Users\hp\Internship\KPU Surabaya\Dataset\DPD\pemilih_disabilitas.xlsx')
```

# 03 Preprocessing

```python
# Melt data pemilih (id = 'uraian')
df_pemilih_melted = df_pemilih.melt(id_vars=['Uraian'], 
                                    var_name='Kecamatan', 
                                    value_name='Jumlah')
df_pemilih_melted.rename(columns={'uraian': 'Uraian'}, inplace=True)

# Melt data surat suara (id = 'Uraian')
df_surat_melted = df_surat.melt(id_vars=['Uraian'], 
                                var_name='Kecamatan', 
                                value_name='Jumlah')

# Melt data disabilitas (id = 'Uraian')
df_disabilitas_melted = df_disabilitas.melt(id_vars=['Uraian'], 
                                            var_name='Kecamatan', 
                                            value_name='Jumlah')
```
